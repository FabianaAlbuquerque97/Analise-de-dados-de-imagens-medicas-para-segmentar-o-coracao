```python
from PIL import Image
```

```python
img = Image.open("la_007.png")
```

```python
cores = []
for cor_rgb in img.getdata():
    if cor_rgb not in cores:
        cores.append(cor_rgb)
```

```python
print(cores)
```

```python
cores = img.convert('RGB').getcolors()
```

```python
print(cores)
```

```python
cores = img.convert('RGB').getcolors(maxcolors=1000)
```

```python
print(cores)
```
